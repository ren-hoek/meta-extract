# Searching with Pymongo

### Import packages and make connection with Mongo client

```python
import pymongo as py
import pprint
from wordcloud import WordCloud
import 
from collections import Counter
import matplotlib.pyplot as plt
%matplotlib inline

client = py.MongoClient('mongo')

db = client['docs']
col = db['aug_meta']
```

### Check whether records have been read in correctly

```python
print("Fail: " + str(col.find({"success": 0}).count()))
print("Success " + str(col.find({"success": 1}).count()))
```

### Print a single record

```python
pprint.pprint(col.find_one())
```

### Search according to a uuid

```python
pprint.pprint(col.find_one({'_id': {ObjectId:'5b3b956c81182b00083a80a3'}}))
```

```python
pprint.pprint(col.find_one({'uuid':'854151e7-2dc5-4d8e-8a2c-e85655284bb7'}))
```

### Generate a word cloud for a specific document

```python
text=col.find_one({'uuid':'854151e7-2dc5-4d8e-8a2c-e85655284bb7'})['content']
```

```python
wordcloud = WordCloud(width=1000, height=750).generate(text)

# Display the generated image:
# the matplotlib way:
plt.figure(figsize=(12,12))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
```

### Return multiple records (in this case the first 5 that failed)

```python
for post in col.find({'success': 0})[:5]:
    print(post)
```

### Repeat for the first 5 that succeeded

```python
for post in col.find({'success': 1})[:5]:
    pprint.pprint(post)
```
